```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_log_error, mean_squared_error, r2_score
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, StackingRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.inspection import permutation_importance
from scipy.stats import pearsonr

# Importing the required libraries for data manipulation, visualisation, and machine learning algorithms
# Pandas is used for data manipulation and handling
# Matplotlib and Seaborn are used for data visualisation
# Numpy is used for numerical operations
# Scikit-learn is used for machine learning algorithms, model evaluation metrics, and data preprocessing
# XGBoost is used for the XGBRegressor model
# Scipy is used for statistical analysis

# Read the data from the file and preprocess it
def read_and_preprocess_data(file_path):
    # Read the CSV file and create a DataFrame
    df = pd.read_csv(file_path)

    # Keep only the relevant columns
    keep_cols = ['Medu', 'Fedu', 'paid', 'health', 'G3']
    # Convert the 'paid' column values from 'no' to 0 and 'yes' to 1
    df['paid'] = df['paid'].replace({'no': 0, 'yes': 1})
    # Remove all columns that are not in the 'keep_cols' list
    df = df.drop(columns=[col for col in df.columns if col not in keep_cols])

    return df

# Perform basic data checks such as missing values, summary statistics, and data types
def basic_data_checks(df):
    print("Missing values:\n", df.isnull().sum())
    print("\nSummary statistics:\n", df.describe())
    print("\nData types:\n", df.dtypes)

# Split the data into training and testing sets, and scale the features
def split_and_scale_data(df, features, target, test_size=0.2, random_state=42):
    # Extract the values of the feature columns and the target column
    X = df[features].values
    y = df[target].values

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    # Scale the features using StandardScaler
    scaler = StandardScaler()
    # Fit and transform the training data
    X_train = scaler.fit_transform(X_train)
    # Transform the testing data
    X_test = scaler.transform(X_test)

    return X_train, X_test, y_train, y_test

# Compute the correlation and p-value between two variables
def compute_correlation_and_pvalue(df, var1, var2):
    # Calculate the Pearson correlation coefficient and p-value
    correlation, p_value = pearsonr(df[var1], df[var2])
    return correlation, p_value

# Create the models to be trained and evaluated
def create_models():
    # Return a dictionary of machine learning models
    return {
        "Dec Tree": DecisionTreeRegressor(random_state=42),
        "RF": RandomForestRegressor(n_estimators=100, random_state=42),
        "xG": XGBRegressor(n_estimators=100, random_state=42),
        "SVM": SVR(kernel='rbf', C=1, gamma='scale'),
        "KNN": KNeighborsRegressor(n_neighbors=5, weights='uniform', metric='euclidean'),
    }

# Train and evaluate the models
def train_evaluate_models(models, X_train, X_test, y_train, y_test):
    # Initialize a dictionary to store the evaluation results
    results = {"Model": [], "MSE": [], "RMSE": [], "MAE": [], "MSLE": [], "R2": []}

    # Train each model and evaluate its performance
    for model_name, model in models.items():
        # Train the model on the training data
        model.fit(X_train, y_train)
        # Make predictions on the test data
        y_pred = model.predict(X_test)

        # Compute evaluation metrics for each model
        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        mae = mean_absolute_error(y_test, y_pred)
        msle = mean_squared_log_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        # Calculate the mean permutation importance
        permutation_importance_mean = compute_permutation_importance(model, X_test, y_test)

        # Append the model name and evaluation metrics to the results dictionary
        results["Model"].append(model_name)
        results["MSE"].append(mse)
        results["RMSE"].append(rmse)
        results["MAE"].append(mae)
        results["MSLE"].append(msle)
        results["R2"].append(r2)

    # Convert the results dictionary to a DataFrame and return it
    return pd.DataFrame(results)

# Perform cross-validation on the models
def cross_validate_models(models, X, y, cv=5):
    # Initialize the results dictionary with keys for the model, mean MSE, and standard deviation of MSE
    results = {"Model": [], "MSE_mean": [], "MSE_std": []}
    
    # Iterate through each model in the input dictionary
    for model_name, model in models.items():
        
        # Create a KFold object with the specified number of splits, shuffling, and a random state
        kf = KFold(n_splits=cv, shuffle=True, random_state=42)
        # Initialize a list to store MSE scores for each fold
        mse_scores = []

        # Iterate through each fold in the KFold object
        for train_index, test_index in kf.split(X):
            # Split the data into training and testing sets using the indices
            X_train, X_test = X[train_index], X[test_index]
            y_train, y_test = y[train_index], y[test_index]
            
            # Fit the model on the training data
            model.fit(X_train, y_train)
            # Predict the target variable for the test data
            y_pred = model.predict(X_test)

            # Calculate the Mean Squared Error for this fold and append it to the mse_scores list
            mse = mean_squared_error(y_test, y_pred)
            mse_scores.append(mse)

        # Append the model name, mean MSE, and standard deviation of MSE to the results dictionary
        results["Model"].append(model_name)
        results["MSE_mean"].append(np.mean(mse_scores))
        results["MSE_std"].append(np.std(mse_scores))

    # Convert the results dictionary to a pandas DataFrame and return it
    return pd.DataFrame(results)

# Visualise the results of the trained models
def visualise_results(results_df):
    # Create a new figure with a specified size
    plt.figure(figsize=(20, 10))

    # Create a subplot for Mean Squared Error comparison
    plt.subplot(2, 3, 1)
    sns.boxplot(x="Model", y="MSE", data=results_df)
    plt.title("Comparison of Mean Squared Error")

    # Create a subplot for Root Mean Squared Error comparison
    plt.subplot(2, 3, 2)
    sns.boxplot(x="Model", y="RMSE", data=results_df)
    plt.title("Comparison of Root Mean Squared Error")

    # Create a subplot for Mean Absolute Error comparison
    plt.subplot(2, 3, 3)
    sns.boxplot(x="Model", y="MAE", data=results_df)
    plt.title("Comparison of Mean Absolute Error")

    # Create a subplot for Mean Squared Log Error comparison
    plt.subplot(2, 3, 4)
    sns.boxplot(x="Model", y="MSLE", data=results_df)
    plt.title("Comparison of Mean Squared Log Error")

    # Create a subplot for R-Squared comparison
    plt.subplot(2, 3, 5)
    sns.boxplot(x="Model", y="R2", data=results_df)
    plt.title("Comparison of R-Squared")

    # Adjust the layout and display the figure
    plt.tight_layout()
    plt.show()
    
# Plot predictions vs actual values for the given models and test data
def plot_predictions_vs_actual(models, X_test, y_test):
    # Create a new figure with multiple subplots, sharing a common y-axis
    fig, axes = plt.subplots(1, 5, figsize=(20, 5), sharey=True)

    # Iterate through the models and the subplots
    for ax, (model_name, model) in zip(axes, models.items()):
        # Predict the target variable for the test data
        y_pred = model.predict(X_test)

        # Create a scatter plot of actual values vs predicted values
        ax.scatter(y_test, y_pred, alpha=0.5)
        # Add a red dashed line to show the ideal 1:1 correlation
        ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
        # Set the title and labels for the subplot
        ax.set_title(f'{model_name} Predictions vs Actual')
        ax.set_xlabel('Actual')
        ax.set_ylabel('Predicted')

    # Adjust the layout and display the figure
    plt.tight_layout()
    plt.show()
    
# Define functions to create various stacking models
def create_stacking_models(base_models, final_estimator):
    stacked_model = StackingRegressor(estimators=base_models, final_estimator=final_estimator)
    return stacked_model

def create_stacked_model_svm_rf(base_models_svm_rf, final_estimator):
    stacked_model_svm_rf = StackingRegressor(estimators=base_models_svm_rf, final_estimator=final_estimator)
    return stacked_model_svm_rf

def create_stacked_model_svm_rf_xg(base_models_svm_rf_xg, final_estimator):
    stacked_model_svm_rf_xg = StackingRegressor(estimators=base_models_svm_rf_xg, final_estimator=final_estimator)
    return stacked_model_svm_rf_xg

def create_stacked_model_svm_rf_xg_dt(base_models_svm_rf_xg_dt, final_estimator):
    stacked_model_svm_rf_xg_dt = StackingRegressor(estimators=base_models_svm_rf_xg_dt, final_estimator=final_estimator)
    return stacked_model_svm_rf_xg_dt

def compute_permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42):
    result = permutation_importance(model, X_test, y_test, n_repeats=n_repeats, random_state=random_state)
    return result.importances_mean

# Display permutation importance for each model
def display_permutation_importance(perm_importance, model_name, features):
    # Create a DataFrame to store permutation importance information for each feature
    perm_importance_df = pd.DataFrame(
        {"Feature": features,
         "Importance Mean": perm_importance.importances_mean,
         "Importance Std": perm_importance.importances_std}
    )
    
    # Add a column with the model's name
    perm_importance_df["Model"] = model_name
    # Rearrange columns in the desired order
    perm_importance_df = perm_importance_df[["Model", "Feature", "Importance Mean", "Importance Std"]]
    return perm_importance_df

# Plot permutation importances for each model
def plot_permutation_importances(perm_importances_df):
    # Create a bar plot to display permutation importances for each feature and model
    plt.figure(figsize=(15, 5))
    sns.barplot(data=perm_importances_df, x="Model", y="Importance Mean", hue="Feature", palette="viridis")
    plt.title("Permutation Importances")
    plt.legend(loc="upper right")
    plt.show()
    
# Rank models based on evaluation metrics
def rank_models(results_df):
    # List of evaluation metrics
    metrics = ["MSE", "RMSE", "MAE", "MSLE", "R2"]

    # Compute the rank of each model for each metric
    rank_df = results_df.copy()
    for metric in metrics:
        # Rank models in descending order for R2
        if metric == "R2":
            rank_df[metric] = results_df[metric].rank(ascending=False)
        # Rank models in ascending order for other metrics
        else:
            rank_df[metric] = results_df[metric].rank(ascending=True)

    # Compute the average rank of each model across all metrics
    rank_df["AvgRank"] = rank_df[metrics].mean(axis=1)

    # Sort the models based on the average rank
    rank_df = rank_df.sort_values("AvgRank")

    return rank_df

def main():
    # Read and preprocess the data
    df = read_and_preprocess_data('math.csv')
    
    # Perform basic data checks
    print("Basic data checks:")
    basic_data_checks(df)
    print("\n")
    
    # List of variable pairs to test for correlation
    hypotheses_vars = [("Medu", "G3"), ("Fedu", "G3"), ("health", "G3"), ("paid", "G3")]

    for var1, var2 in hypotheses_vars:
        correlation, p_value = compute_correlation_and_pvalue(df, var1, var2)
        print(f"Correlation between {var1} and {var2}: {correlation:.2f}, p-value: {p_value:.4f}")

    # Select features and target
    features = ["Medu", "Fedu", "health", "paid"]
    target = "G3"

    # Split and scale the data
    X_train, X_test, y_train, y_test = split_and_scale_data(df, features, target)
    # Create models
    models = create_models()
    
    # Perform cross-validation on the models
    cv_results_df = cross_validate_models(models, X_train, y_train)

    # Train and evaluate the models
    results_df = train_evaluate_models(models, X_train, X_test, y_train, y_test)
    print("\nModel Performance:\n", results_df)
    
    # Calculate and display permutation importances
    perm_importances = []
    for model_name, model in models.items():
        perm_importance = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42)
        perm_importances.append(display_permutation_importance(perm_importance, model_name, features))
    
    perm_importances_df = pd.concat(perm_importances, ignore_index=True)
    print("\nPermutation Importances:\n", perm_importances_df)
    
    # Plot permutation importances
    plot_permutation_importances(perm_importances_df)

    # Visualize the results
    visualise_results(results_df)
    
    # Plot predictions vs actual values
    plot_predictions_vs_actual(models, X_test, y_test)

    # Initialize the base models for stacking
    base_models = [
        ("Decision Tree", DecisionTreeRegressor(random_state=42)),
        ("Random Forest", RandomForestRegressor(n_estimators=100, random_state=42)),
        ("Gradient Boosting", XGBRegressor(n_estimators=100, random_state=42)),
        ("Support Vector Regression", SVR(kernel='rbf', C=1, gamma='scale')),
        ("K-Nearest Neighbors", KNeighborsRegressor(n_neighbors=5, weights='uniform', metric='euclidean'))
    ]
    
    # Train, predict and evaluate each individual model
    models = create_models()
    results_df = train_evaluate_models(models, X_train, X_test, y_train, y_test)
    
    # Print cross-validation results
    print("Cross-validation results:\n", cv_results_df)

    # Rank the models based on the metrics
    ranked_models = rank_models(results_df)
    print("\nIndividual Models Ranking:\n", ranked_models)

    # Create and train a stacking model
    stacked_model = create_stacking_models(base_models, LinearRegression())
    stacked_model.fit(X_train, y_train)
    y_pred = stacked_model.predict(X_test)

    # Evaluate the stacked model
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    msle = mean_squared_log_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # Initialize the base models with only SVM and RF
    base_models_svm_rf = [
        ("SVM", SVR(kernel='rbf', C=1, gamma='scale')),
        ("Random Forest", RandomForestRegressor(n_estimators=100, random_state=42))
    ]

    # Initialize the stacking regressor with only SVM and RF
    stacked_model_svm_rf = create_stacked_model_svm_rf(base_models_svm_rf, LinearRegression())
    
    # Initialize the base models with only SVM, RF and xG
    base_models_svm_rf_xg = [
        ("SVM", SVR(kernel='rbf', C=1, gamma='scale')),
        ("Random Forest", RandomForestRegressor(n_estimators=100, random_state=42)),
        ("Gradient Boosting", XGBRegressor(n_estimators=100, random_state=42))
    ]

    # Initialize the stacking regressor with only SVM, RF and xG
    stacked_model_svm_rf_xg = create_stacked_model_svm_rf_xg(base_models_svm_rf_xg, LinearRegression())
    
    # Initialize the base models with SVM, RF, xG and DT
    base_models_svm_rf_xg_dt = [
        ("SVM", SVR(kernel='rbf', C=1, gamma='scale')),
        ("Random Forest", RandomForestRegressor(n_estimators=100, random_state=42)),
        ("Gradient Boosting", XGBRegressor(n_estimators=100, random_state=42)),
        ("Decision Tree", DecisionTreeRegressor(random_state=42))
    ]

    # Initialize the stacking regressor with SVM, RF, xG and DT
    stacked_model_svm_rf_xg_dt = create_stacked_model_svm_rf_xg_dt(base_models_svm_rf_xg_dt, LinearRegression())

    # Create a dictionary to compare SVM with Stacked, SVM+RF, SVM+RF+xG and  SVM+RF+xG+DT.
    models = {
        "SVM": SVR(kernel='rbf', C=1, gamma='scale'),
        "Stacked": stacked_model,
        "SVM+RF": stacked_model_svm_rf,
        "SVM+RF+xG": stacked_model_svm_rf_xg,
        "No KNN": stacked_model_svm_rf_xg_dt,
    }

    # Train, predict, and evaluate the models
    results_df = train_evaluate_models(models, X_train, X_test, y_train, y_test)
    print("\nStacked Models v SVM:\n", results_df)
    
    # Rank the models based on the metrics
    ranked_models = rank_models(results_df)
    print("\nStacked Models v SVM Ranking:\n", ranked_models)

    # Visualize the results
    visualise_results(results_df)
    
    # Plot predictions vs actual values
    plot_predictions_vs_actual(models, X_test, y_test)

if __name__ == "__main__":
    main()
```
